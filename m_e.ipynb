```python
import pandas as pd

# M&E စက်ပစ္စည်းတွေရဲ့ Data လေးတစ်ခု ဖန်တီးကြည့်ရအောင်
data = {
    'Device': ['Generator', 'Chiller 1', 'Chiller 2', 'UPS'],
    'Status': ['Running', 'Running', 'Down', 'Running'],
    'Temperature': [65, 70, 0, 45]
}

# ဒီ Data ကို DataFrame (Excel ပုံစံ) အဖြစ် ပြောင်းလိုက်မယ်
df = pd.DataFrame(data)

# ရလာတဲ့ Table ကို ပြမယ်
print(df)
```

```python
# 'Status' က 'Down' ဖြစ်နေတဲ့ row တွေကိုပဲ ခွဲထုတ်မယ်
faulty_devices = df[df['Status'] == 'Down']

# ရလဒ်ကို ပြမယ်
print("Faulty Devices Found:")
print(faulty_devices)
```

```python
import pandas as pd

# M&E အတွက် စမ်းသပ် Data လေး
data = {
    'Device': ['Generator', 'Chiller 1', 'Chiller 2', 'UPS', 'AC Unit'],
    'Status': ['Running', 'Running', 'Down', 'Running', 'Down'],
    'Temperature': [65, 70, 0, 45, 80]
}

# Dataframe တည်ဆောက်မယ်
df = pd.DataFrame(data)
print(df)

# 'my_me_report.xlsx' ဆိုတဲ့ နာမည်နဲ့ Excel ဖိုင်အဖြစ် သိမ်းဆည်းမယ်
df.to_excel('my_me_report.xlsx', index=False)

print("Excel ဖိုင် 'my_me_report.xlsx' ကို အောင်မြင်စွာ ဖန်တီးပြီးပါပြီ။")
```

```python
import pandas as pd

# အခုလေးတင် ဖန်တီးထားတဲ့ Excel ဖိုင်ကို ပြန်ဖတ်မယ်
df_read = pd.read_excel('my_me_report.xlsx')

# 'Status' က 'Down' ဖြစ်နေတဲ့ စက်တွေကို ခွဲထုတ်မယ်
faulty_devices = df_read[df_read['Status'] == 'Down']

# ရလဒ်ကို ပြမယ်
print("--- အရေးပေါ်စစ်ဆေးရန် လိုအပ်သော စက်များ ---")
print(faulty_devices)
```

```python
# အရေးပေါ် စာရင်းကို Excel ဖိုင်အသစ်တစ်ခုအနေနဲ့ သိမ်းမယ်
faulty_devices.to_excel('Faulty_Report.xlsx', index=False)

print("Report အသစ် 'Faulty_Report.xlsx' ကို ဖန်တီးပြီးပါပြီ။")
```

```python
import matplotlib.pyplot as plt

# Status တစ်ခုချင်းစီ ဘယ်နှစ်ခုရှိလဲ ရေတွက်မယ်
status_counts = df['Status'].value_counts()

# ဂရပ်ဆွဲမယ်
status_counts.plot(kind='bar', color=['green', 'red'], figsize=(8, 5))

# ဂရပ်ကို အလှဆင်မယ်
plt.title('Device Status Report')
plt.xlabel('Status')
plt.ylabel('Count')
plt.xticks(rotation=0) # စာသားတွေကို မစောင်းအောင်ထားမယ်
plt.show()
```
